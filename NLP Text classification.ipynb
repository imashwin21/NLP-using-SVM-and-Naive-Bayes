# SVM

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("C:/Users/imash/Downloads/nlp_dataset.csv")
df = pd.DataFrame(df)
df
```

```python
x = df['Comment']
y = df["Emotion"]
```

```python
x
```

```python
y
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x1 = cv.fit_transform(x)
```

```python
x1
```

```python
x1.toarray()
```

```python
cv.get_feature_names_out()
```

```python
import pandas as pd
pd.DataFrame(x1.toarray(),columns=cv.get_feature_names_out())
```

```python
x_train, x_test, y_train, y_test = train_test_split(x1,y, test_size=0.2)

#x1 is sparse matrix
```

```python
x_test
```

```python
x_train
```

As the problem is Supervised Machine Learning classification problem, we can choose Logistic regression model

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train,y_train)
```

```python
y_pred = svc.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
sentence1 = ['She is very annoying']
emotion1 = cv.transform(sentence1)
emotion1
```

```python
svc.predict(emotion1)
```

```python
sentence2 = ["she is a very good person"]
emotion2 = cv.transform(sentence2)
svc.predict(emotion2)
```

# Naive Bayes

Naive Bayes is a simple probabilistic classifier that predicts the category of an item based on the likelihood of its feautures relationship.

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB()
mnb.fit(x_train,y_train)
```

```python
y_pred = mnb.predict(x_test)
accuracy_score(y_test,y_pred)
```
